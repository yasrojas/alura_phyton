# Introdução da biblioteca seaborn com matpotlib. Utilização de histogramas, botplox e countplot. 

```python
mapa = {'droga': 'composto'}
dados.rename(columns = mapa, inplace= True)
```

```python
dados.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))
plt.grid(True)

sns.set()
ax = sns.countplot(x = 'composto', 
                   data = dados.query('composto in @cod_compostos'), 
                   order=dados.query('composto in @cod_compostos').composto.value_counts().index) 
ax.set_title('Top 5 Compostos', fontsize='xx-large', fontweight='bold', pad= 30)
ax.set_xlabel('tipo de composto', fontsize='large', fontstyle= 'italic')
ax.set_ylabel('quantidade de composto', fontsize='large', fontstyle= 'italic')

plt.show()
```

```python
# dados.loc[:, 'g_0':'g_771'].describe().T['mean'].hist(bins=50)

ax = sns.histplot(data = dados, 
                  x = dados.loc[:, 'g_0':'g_771'].describe().T['mean'], 
                  bins = 50,
                  kde= True)
ax.set_title('Histograma da média de g-0 ao g-771', 
             fontsize='xx-large', 
             fontweight='bold', 
             pad= 30)
ax.set_xlabel('valores em g-n', fontsize='large')
ax.set_ylabel('quantidade do valor nos dados', fontsize='large')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.set_theme(style="whitegrid")
ax = sns.boxplot(x= 'g_0', 
                 y = 'tratamento', 
                 data = dados)
ax.set_title('Botplox da relação entre g-0 e tratamento', 
             fontsize='xx-large', 
             fontweight='bold', 
             pad= 30)
ax.set_xlabel('valores em g-0', fontsize='large')
ax.set_ylabel('tipo de tratamento', fontsize='large')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.set_theme(style="whitegrid")
ax = sns.boxplot(x = 'composto', 
                 y = 'g_0', 
                 data = dados.query('composto in @cod_compostos'))
ax.set_title('Botplox da relação entre g-0 e principais compostos', 
             fontsize='xx-large', 
             fontweight='bold', 
             pad= 30)
ax.set_xlabel('tipo de composto', fontsize='large')
ax.set_ylabel('valor de g-0', fontsize='large')

plt.show()
```
